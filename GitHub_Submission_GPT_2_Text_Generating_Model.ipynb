## Download GPT-2

There are three released sizes of GPT-2:

* `124M` (default): the "small" model, 500MB on disk.
* `355M`: the "medium" model, 1.5GB on disk.
* `774M`: the "large" model
* `1558M`: the "extra large", true model. 

Larger models have more knowledge, but take longer to finetune and longer to generate text. You can specify which base model to use by changing `model_name` in the cells below.


```
gpt2.download_gpt2(model_name="124M")
```

## Finetune GPT-2

The next cell will start the actual finetuning of GPT-2. To have the finetuning run indefinitely, set `steps = -1`

The model checkpoints will be saved in `/checkpoint/run1` by default. 

Other optional-but-helpful parameters for `gpt2.finetune`:


```
sess = gpt2.start_tf_sess()

gpt2.finetune(sess,
              dataset="moby_dick_full text.txt",
              model_name='124M',
              steps=200,
              restore_from='fresh',
              run_name='run1',
              print_every=9,
              sample_every=20,
              save_every=20
              )
```
